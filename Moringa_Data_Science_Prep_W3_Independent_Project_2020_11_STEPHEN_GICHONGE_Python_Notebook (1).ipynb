# Data Preparation

## Importing libraries

```
#Import the pandas library
import pandas as pd
# as well as the Numpy library
import numpy as np
```

## Loading Datasets

```
# Loading a cell_geo csv
df = pd.read_csv('cells_geo.csv')
df.head(10)

```

### Cleaning the cells_geo Datataset

```
# split the table into columns
df = df[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].apply(lambda y: pd.Series(y.split(';')))
df.head(10)

```

```
#removing the quotes from the data
for r, col in enumerate(df.columns):
    df.iloc[:, r] = df.iloc[:, r].str.replace('"', '')
df.head(10)
```

```
#dropping the duplicate indexes
df_new = df.drop([0], axis = 1)
df_new
```

```

#Name the columns(changing side_id to site_code to match records in other datasets)
df_new.columns = ['VILLES','STATUS', 'LOCALISATION', 'DECOUPZONE', 'ZONENAME', 'LONGITUDE', 'LATITUDE', 'REGION', 'AREA', 'CELL_ID', 'SITE_ID']
df_new.head(5)
```

```
#check for duplicated records
df_new.duplicated()
```

### Cleaning the Telcom Datatasets

```
# Loading telcom csv
df1 = pd.read_csv('Telcom_dataset.csv')
df1.head(10)

# Loading a telcom2 csv
df2 = pd.read_csv('Telcom_dataset2.csv')
df2.head(10)

# Loading a telcom csv
df3 = pd.read_csv('Telcom_dataset3.csv')
df3.head(10)
```

```
#changing 'PRODUTC' to 'PRODUCT' and 'DATETIME' to 'DATE_TIME'
df1 = pd.read_csv('Telcom_dataset.csv')
df1.head(5)
df1.rename(columns = {'PRODUTC':'PRODUCT', 'DATETIME':'DATE_TIME' }, inplace = True) 
df1.head(5)

#changing 'CELLID' to 'CELL_ID' and 'SIET_ID' to 'SITE_ID'
df3 = pd.read_csv('Telcom_dataset3.csv')
df3.head(5)
df3.rename(columns = {'CELLID':'CELL_ID', 'SIET_ID':'SITE_ID' }, inplace = True) 
df3.head(5)

```

```
#merge all telcom datasets based on #DATE_TIME'
#first telcom1 and telcom2
df_new1 = pd.merge(df1,df2, how='outer', on=['SITE_ID', 'DATE_TIME', 'PRODUCT', 'VALUE','CELL_ID'])
df_new1
df_new1.head(5)

#secondly, telcom1,2 and telom3
df_new2 = pd.merge(df_new1,df3, how='outer', on=['SITE_ID', 'DATE_TIME', 'PRODUCT', 'VALUE','CELL_ID'])
df_new2
df_new2.head(5)

#drop duplicates
df_new2.duplicated()
df_c = df_new2.drop_duplicates()
df_c

#removing any null records
df_cleaned = df_c.dropna(how='all')
df_cleaned
#df_cleaned.head(2)

```

```
#drop unwanted columns

df_cleaned.drop(['CELL_ON_SITE','CELL_ON_SITE_x', 'CELL_ON_SITE_y','COUNTRY_A', 
                 'COUNTRY_B', 'COUNTRY_A_y', 'COUNTRY_B_y', 'DW_A_NUMBER','DW_B_NUMBER',
                 'DW_A_NUMBER_INT_x','DW_B_NUMBER_INT_x', 'DW_A_NUMBER_INT_y','DW_B_NUMBER_INT_y',
                 'COUNTRY_A_x', 'COUNTRY_B_x'], axis=1, inplace=True)

df_cleaned
```

```
#merge cell_geo and telcom datasets based on site id

df_all = pd.merge(df_new,df_cleaned, how='left', on=['SITE_ID'])
df_all
#df_all.head(2)

#dropping of duplicates based on columns date_time, product and value
df_all1 = df_all.drop_duplicates(['DATE_TIME','PRODUCT', 'VALUE'], keep='last')
df_all1

#resetting the indexes
df_all1.reset_index(inplace = True) 
df_all1

#dropping any records with nulls
df_all2 = df_all1.dropna(how='all')
df_all2.head(5)


#drop unwanted columns
df_all2.drop(['index', 'CELL_ID_x', 'CELL_ID_y'], axis=1, inplace=True)
df_all2.head(5)
#df_all2
```

# Data Analysis

## Check Status of customers

```
#before carrying out analysis we need to check that within our cleaned records
active_list = ['In Service']
df_all2[df_all2.STATUS.isin(active_list)]

#4441 records of the 4524 are active the remaining have empty status
active_list = ['In Service']
df_all2[~df_all2.STATUS.isin(active_list)]

#we need to fill the missing status as Not In Service
df_all2['STATUS'].replace('', ' Not In Service', inplace=True) 
df_all2[~df_all2.STATUS.isin(active_list)]

```

## Location Analysis

```
#observe sites(unique) with most activity based on value
#groupby villes, sites, product and value

grps = df_all2.groupby(['VILLES', 'SITE_ID','PRODUCT'])['VALUE'].sum()
grps

#producing a table to reflect the above(value aggregated)
s = pd.pivot_table(df_all2, index=['VILLES', 'SITE_ID'], aggfunc='sum')
s


```

```
#observe cities(villes) with most activity based on value
#groupby villes, product and value(aggregate)

grpc = df_all2.groupby(['VILLES', 'PRODUCT'])['VALUE'].sum()
grpc

#producing a table to reflect the above
y = pd.pivot_table(df_all2, index=['VILLES'], aggfunc='sum')
y
```

```
# Number of observations used by category(checking sample size)
df_all2.groupby(['VILLES', 'PRODUCT']).size()

# Average value of Voice and sms by VILLE
df_all2['VALUE'].groupby([df_all2['VILLES'], df_all2[ 'PRODUCT']]).mean()

#producing a table to reflect the above
vs = pd.pivot_table(df_all2, index=['VILLES', 'PRODUCT'], aggfunc='mean')
vs
```

```
#observe zones with most activity based on value
#groupby zones, villes, product and value(aggregate)

grpz = df_all2.groupby(['ZONENAME','VILLES', 'PRODUCT'])['VALUE'].sum()
grpz

#producing a table to reflect the above
gz = pd.pivot_table(df_all2, index=['ZONENAME','VILLES'], aggfunc='sum')
gz
```

```
#observe localisations with most activity based on value
#groupby localisations,villes, product and value

grpl = df_all2.groupby(['LOCALISATION','ZONENAME','VILLES','PRODUCT'])['VALUE'].sum()
grpl

#producing a table to reflect the above
gl = pd.pivot_table(df_all2, index=['LOCALISATION','ZONENAME'], aggfunc='sum')
gl

#showing value by localisation alone
gl1 = pd.pivot_table(df_all2, index=['LOCALISATION'], aggfunc='sum')
gl1
```

## Duration related Analysis

```
#separating the date from time
#converting the date to datetime format but first float
#checking data types
df_all2.dtypes

df_all2['datetime']=pd.to_datetime(df_all2['DATE_TIME'], dayfirst=True) 
df_all2.datetime.head()

```

```
#converting time to mins
df_all2['MINS'] = df_all2['datetime'].dt.hour * 60 + df_all2['datetime'].dt.minute + df_all2['datetime'].dt.second/60
df_all2

#converting time to float
df_all2['datetime'] = df_all2['datetime'].values.astype(float)
df_all2

#checking data types again
df_all2.dtypes
```
